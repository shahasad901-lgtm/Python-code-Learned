```python
 # WRITE A PRO TO CHECK IF A NUMBER IS  DIVEDABLE:
def check_divisible(number, divider):
    if number % divider == 0:
        print(f"{number}, {divider} se poora divide hota hai!")
    else:
        print(f"{number}, {divider} se divide nahi hota.")

# Check karte hain: Kya 10, 2 se divide hota hai?
check_divisible(10, 2)
```

```python
# WRITE A PROGRAM TO COUNT NUMBERS FROM 1 -20  AND PRINT ONLH MULTIPLES OF 3
def count_threes():
    for i in range(1, 21): # 1 se shuru, 21 se pehle khatam
        if i % 3 == 0:     # Agar 3 se divide ho jaye
            print(i)

count_threes()
```

```python
# WAP TO REVERSE A NUMBER
def ulta_number(n):
    text = str(n)        # Number ko text banaya "123"
    reverse = text[::-1] # Python ka magic trick ulta karne ke liye
    print(f"Asli: {n}, Ulta: {reverse}")

ulta_number(543)
```

```python
# WAP TO FIND THE SUM OF DIGIT OF A MUMBER
def sum_of_digits(n):
    total = 0
    text_n = str(n)         # Number ko text banaya "123"
    for digit in text_n:    # Ek ek karke har number (1, 2, 3) par jayega
        total += int(digit) # Text ko wapis number bana kar total mein jama (add) kiya
    print(f"Digits ka total: {total}")

sum_of_digits(123)
```

```python
# WAP TO COUNT VOWELS IN A STRING
def count_vowels(word):
    vowels = "aeiouAEIOU"
    count = 0
    for char in word:       # Word ke har harf (letter) ko check karo
        if char in vowels:  # Agar wo vowel hai
            count += 1      # Ginti mein ek izafa karo
    print(f"Vowels ki tadad: {count}")

count_vowels("Pakistan")
```

```python
# WAP TO CHECK IF A WORD STARTS WATH VOWEL
def starts_with_vowel(word):
    vowels = "aeiouAEIOU"
    first_letter = word[0] # Pehla harf pakra
    if first_letter in vowels:
        print(f"Yes! '{word}' vowel se shuru hota hai.")
    else:
        print(f"No! '{word}' vowel se shuru nahi hota.")

starts_with_vowel("Apple") 
starts_with_vowel("Cat")

```

```python
# WAP TO FIND LARGEST NUMBER IN THE LIST
def find_largest(numbers):
    # Python ka built-in function 'max' sab kaam asaan kar deta hai
    bada_number = max(numbers)
    print(f"Sab se bada number: {bada_number}")

my_list = [10, 85, 42, 90, 31]
find_largest(my_list)
```

```python
# WAP TO REMOVE DUPLICATE NYUMBER FROM THE LIST
def remove_duplicates(my_list):
    # Python mein 'set' ka matlab hai "Sirf Unique Cheezain"
    unique_list = list(set(my_list))
    print(f"Baghair duplicates ke: {unique_list}")

numbers = [10, 20, 20, 30, 40, 10, 50]
remove_duplicates(numbers)
```

```python
# WAP TO COUNT CHARACTER IN A STRING
def count_chars(text):
    length = len(text) # 'len' matlab Length
    print(f"Total characters: {length}")

count_chars("Python Coding")
```

```python
# WAP TO COUNT HOW MANY NMBERS IN A LIST  GREATER THAN 10
def count_greater_than_10(numbers):
    count = 0
    for n in numbers:
        if n > 10:
            count += 1
    print(f"10 se baray numbers ki tadad: {count}")

my_numbers = [5, 12, 8, 20, 3, 15]
count_greater_than_10(my_numbers) 
```

```python
# WAP TO FIND OUT THE AVERAGE OF NUMBER
def find_average(numbers):
    total_sum = sum(numbers)      # Sab ko jama kiya
    total_count = len(numbers)    # Gina ke kitne hain
    average = total_sum / total_count
    print(f"Average hai: {average}")

find_average([10, 20, 30, 40, 50])
```
