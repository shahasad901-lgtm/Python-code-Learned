```python
# Write a Python program to create a class Person with a method show_name() that prints the Person’s name.
class Person:
    def __init__(self, name):
        self.naam = name

    def show_name(self):
        print(f"Is shakhs ka naam {self.naam} hai.")

p1 = Person("Ali")
p1.show_name()
# Logic: Class ke andar jo function hota hai usay "Method" kehte hain. 
#     Method hamesha self ko pehle leta hai taake usay pata ho kis bande ka naam dikhana hai.
```

```python
# Write a Python program to create a class Rectangle with attributes length and width and print the area.
class Rectangle:
    def __init__(self, length, width):
        self.l = length
        self.w = width

    def show_area(self):
        area = self.l * self.w
        print(f"Rectangle ka Area: {area}")

r1 = Rectangle(10, 5)
r1.show_area()
# Logic: Area ka formula Length * Width hota hai. Humne __init__ mein values save ki aur method mein multiply kar diya.
```

```python
# Write a Python program to create a class Circle and calculate the area of the circle.
class Circle:
    def __init__(self, radius):
        self.r = radius

    def calculate_area(self):
        # Formula: 3.14 * r * r
        area = 3.14 * (self.r * self.r)
        print(f"Circle ka Area: {area}")

c1 = Circle(7)
c1.calculate_area()
# Logic: Circle ke liye humein sirf radius (r) chahiye hota hai.
```

```python
# Write a Python program that creates a class with a destructor that prints
# # "Object destroyed".
class Demo:
    def __del__(self):
        print("Object destroyed")

d1 = Demo()
del d1
# Logic: __init__ paidaish ke liye hai, __del__ khatam hone ke liye.
# Jab hum del keyword use karte hain, toh Python memory se us object ko nikaal deta hai.
```
